## Building a Corpus of Documents

```python
# define a list of sentences and label them with a category
corpus = ['The sky is blue and beautiful.',
          'Love this blue and beautiful sky!',
          'The quick brown fox jumps over the lazy dog.',
          "A king's breakfast has sausages, ham, bacon, eggs, toast and beans",
          'I love green eggs, ham, sausages and bacon!',
          'The brown fox is quick and the blue dog is lazy!',
          'The sky is very blue and the sky is very beautiful today',
          'The dog is lazy but the brown fox is quick!'    
]
labels = ['weather', 'weather', 'animals', 'food', 'food', 'animals', 'weather', 'animals']

corpus = np.array(corpus)
corpus_df = pd.DataFrame({'Document': corpus, 
                          'Category': labels})
corpus_df = corpus_df[['Document', 'Category']]
corpus_df
```

```python
corpus_df.info()
```

## Preprocessing the corpus

Examples of this include

- Removing HTML tags

- Removing Accented Characters

- Expanding contractions

- Stemming and lemmatization

- Removing stopwords

```python
wpt = nltk.WordPunctTokenizer()
stop_words = nltk.corpus.stopwords.words('english')

def normalize_document(doc):
    # remove any character that isn't alphabetic or whitespace
    doc = re.sub(r'[^a-zA-Z\s]', '', doc, re.I|re.A)
    # make the sentence lowercase
    doc = doc.lower()
    # Remove whitespace from the beginning and end of lines
    doc = doc.strip()
    # tokenize document
    tokens = wpt.tokenize(doc)
    # filter stopwords out of document
    filtered_tokens = [token for token in tokens if token not in stop_words]
    # re-create document from filtered tokens
    doc = ' '.join(filtered_tokens)
    return doc

# The makes the function able to work with arrays/vectors from numpy rather than defining a wrapper function ourselves
normalize_corpus = np.vectorize(normalize_document)

corpus_norm = normalize_corpus(corpus)
corpus_norm
```

## Traditional Feature Engineering Models

Now that we have pre-processed our corpus, we can perform some feature extraction!

In this section we will use count-based strategies such as the Bag of Words model. These kind of models don't take word order into account

### Bag of Words Model

Represents each document as a vector, where each dimension is a word, and the value could be something like the frequency within the document or simply whether the word is present.

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(min_df=0., max_df=1.)

# stored as a sparse matrix in order to save space
cv_matrix = cv.fit_transform(corpus_norm)
cv_matrix
```

```python
print(cv_matrix)
```

```python
# We can also view the matrix in a dense representation, although this uses much more memory
cv_matrix = cv_matrix.toarray()
cv_matrix
```

```python
# A more visual representation is that of a pandas dataframe, which will show us what each of the columns represent

# unique words in the corpus
vocab = cv.get_feature_names_out()

corpus_vocab = pd.DataFrame(cv_matrix, columns=vocab)
corpus_vocab
```

### Bag of N-Grams

This is like bag of words, except we're using ngrams, which are phrases that are a certain number of words. In this example, we will use bigrams.

```python
# Make it easy to demonstrate vectorizers
def extract_features(corpus, extractor):
    f_mat = extractor.fit_transform(corpus).toarray()
    vocab = extractor.get_feature_names_out()
    return pd.DataFrame(np.round(f_mat, 2), columns=vocab)
```

```python
bv = CountVectorizer(ngram_range=(2,2))

extract_features(corpus_norm, bv)
```

```python
bv = CountVectorizer(ngram_range=(1,3))

extract_features(corpus_norm, bv)
```

## TF-IDF Model

TF-IDF, or term frequency-inverse document frequency fixes some of the flaws of the bag of words model0

### TfidfTransformer

Transforms an existing bag of words into TF-IDF form

```python
from sklearn.feature_extraction.text import TfidfTransformer

tt = TfidfTransformer(norm='l2', use_idf=True)

vocab = cv.get_feature_names_out()
extract_features(cv_matrix, tt).set_axis(vocab, axis=1, inplace=False)
```

### TfidfVectorizer

This is exactly like the TfidfTransformer, except it will also perform the term frequency computation

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer(min_df=0., max_df=1., norm='l2', use_idf=True, smooth_idf=True)

extract_features(corpus_norm, tv)
```

### TF-IDF Behind the Scenes

```python
# Unique words as feature names
words = [word for doc in [doc.split() for doc in corpus_norm] for word in doc]
unique_words = list(set(words))

default_feature_dict = { w: 0 for w in unique_words }

print('Feature Names: ', unique_words)
```

```python
# Counter is a data structure that takes elements as the keys and the number
# of occurences as the values
from collections import Counter

# build a bag of words for each document using term frequencies
bow_features = []
for doc in corpus_norm:
    bow_feature_doc = Counter(doc.split())
    all_features = Counter(default_feature_dict)
    bow_feature_doc.update(all_features)
    bow_features.append(bow_feature_doc)

bow_features = pd.DataFrame(bow_features)
bow_features
```

```python
import scipy.sparse as sp
feature_names = list(bow_features.columns)

# compute the document frequencies, or how many times the word shows up overall
df = np.diff(sp.csc_matrix(bow_features, copy=True).indptr)
df = 1 + df

pd.DataFrame([df], columns=feature_names)
```

```python
# now we need the inverse document frequencies

total_docs = 1 + len(corpus_norm)
idf = 1.0 + np.log(float(total_docs) / df)

pd.DataFrame([np.round(idf, 2)], columns=feature_names)
```

```python
# now that we have the smoothed document frequencies, let's convert them into a matrix

total_features = bow_features.shape[1]
idf_diag = sp.spdiags(idf, diags=0, m=total_features, n=total_features)
idf_dense = idf_diag.todense()

pd.DataFrame(np.round(idf_dense, 2))
```

```python
# compute the feature matrix
tf = np.array(bow_features, dtype='float64')
tfidf = tf * idf

pd.DataFrame(np.round(tfidf, 2), columns=feature_names)
```

```python
# We have the feature matrix, but one last thing we need to do is divide it by the L2 norm.

from numpy.linalg import norm

norms = norm(tfidf, axis=1)

print(np.round(norms, 3))

# compute the normalized tfidf
norm_tfidf = tfidf / norms[:, None]

pd.DataFrame(np.round(norm_tfidf, 2), columns=feature_names)
```

## Document Similarity

This is the process of using a distance or other similarity based metric that can determine how similar a text document is to other documents based on features that were extracted from the documents.

This is useful in domains such as search engines, document clustering, and information retrieval.

There are several similarity measures that can be used to compare documents such as

- cosine similarity
- distances (euclidean, manhattan)
- BM25 similarity
- jaccard distance
- and more

```python
from sklearn.metrics.pairwise import cosine_similarity

tv_matrix = tv.fit_transform(corpus).toarray()
similarity_matrix = cosine_similarity(tv_matrix)
similarity_df = pd.DataFrame(similarity_matrix)
similarity_df
```

### Document Clustering with Similarity Features

Agglomerative hierarchial clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(similarity_matrix, 'ward')
pd.DataFrame(Z, columns=['Document/Cluster 1', 'Document/Cluster 2', 'Distance', 'Cluster Size'], dtype='object')
```

```python
plt.figure(figsize=(8,3))
plt.title('Hierarchial Clustering Dendogram')
plt.xlabel('Data Point')
plt.ylabel('Distance')
dendrogram(Z)
plt.axhline(y=1.0, c='k', ls='--', lw=0.5)
```

```python
from scipy.cluster.hierarchy import fcluster
max_dist = 1.2
cluster_labels = fcluster(Z, max_dist, criterion='distance')
cluster_labels = pd.DataFrame(cluster_labels, columns=['ClusterLabel'])
pd.concat([corpus_df, cluster_labels], axis=1).sort_values('ClusterLabel')
```

## Topic Models

Topic modeling is the process of extracting themes or ideas from a corpus of documents. Each topic can be represented as a bag of words from the corpus.

Topic models are extremely useful in summarizing large corpuses of text documents to extract key concepts or extracting features that capture patterns in the data.

```python
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components=3, max_iter=10000, random_state=0)

dt_matrix = lda.fit_transform(cv_matrix)
features = pd.DataFrame(dt_matrix, columns=['T1', 'T2', 'T3'])
features
```

```python
tt_matrix = lda.components_

topics = pd.DataFrame(columns=vocab)
for topic_weights in tt_matrix:
    topic = {token: weight for token, weight in zip(vocab, topic_weights)}
    topics = topics.append(topic, ignore_index=True)

for index, row in topics.iterrows():
    print(row.where(lambda x: x > 0.6).dropna())
```
